```python
from pathlib import Path
import pandas as pd

cars = pd.read_csv(Path("car_price_prediction.csv"))
cars.head()
```

```python
cars.info()
```

```python
cars = cars.fillna(0)
df = pd.DataFrame(cars)
df = df.drop('ID', axis=1)
df = df.drop('Levy', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Manufacturer = le.fit_transform(df['Manufacturer'])
Model = le.fit_transform(df['Model'])
Category = le.fit_transform(df['Category'])
Leather_interior = le.fit_transform(df['Leather interior'])
Fuel_type = le.fit_transform(df['Fuel type'])
Engine_volume = le.fit_transform(df['Engine volume'])
Mileage = le.fit_transform(df['Mileage'])
Gear_box_type = le.fit_transform(df['Gear box type'])
Drive_wheels = le.fit_transform(df['Drive wheels'])
Doors = le.fit_transform(df['Doors'])
Wheel = le.fit_transform(df['Wheel'])
Color = le.fit_transform(df['Color'])

Manufacturer, Model , Category, Leather_interior, Fuel_type, Engine_volume, Mileage, Gear_box_type, Drive_wheels, Doors, Wheel, Color
```

```python
df['NManufacturer'] = Manufacturer
df['NModel'] = Model
df['NCategory'] = Category
df['NLeather interior'] = Leather_interior
df['NFuel type'] = Fuel_type
df['NEngine volume'] = Engine_volume
df['NMileage'] = Mileage
df['NGear box type'] = Gear_box_type
df['NDrive wheels'] = Drive_wheels
df['NDoors'] = Doors
df['NWheel'] = Wheel
df['NColor'] = Color
new_dataset = df.drop(columns=['Manufacturer',
                  'Model',
                  'Category',
                  'Leather interior',
                  'Fuel type',
                  'Engine volume',
                  'Mileage',
                  'Cylinders',
                  'Gear box type',
                  'Drive wheels',
                  'Doors',
                  'Wheel',
                  'Color'])
new_dataset
```

```python
new_dataset.to_csv('Cars.csv', index=False)
```
